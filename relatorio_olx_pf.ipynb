```python
import pandas as pd
import numpy as np
import re
from datetime import datetime

olx = pd.read_excel('Movida_Vendedores.xlsx')
email_loja = pd.read_excel('email_loja.xlsx')
propostas = pd.read_csv('propostas.csv', sep=";", low_memory=False)

olx_pf = olx.copy()

# Eliminar colunas desnecessárias
olx_pf = olx_pf.drop(columns=['Grupo','Nome da Conta', 'Validação da Conta',
       'Validação do Telefone', 'Chat ou Form. Email',
       'Login no app',  'Dispositivo Acessado',
       'Início do Plano', 'Plano',
        'Tempo Médio de Resposta', 'Cliques WhatsApp',
       'Cliques Telefone',  'Média de Anúncios',
       'Lead por Anúncio'])

# Junção com planilha email_loja
olx_pf = olx_pf.merge(email_loja, on='E-mail', how='left')

# Transformar valores NaN de 'Account ID' em 0
olx_pf['Account ID'] = olx_pf['Account ID'].fillna(0)

# Transformar valores NaN de 'Leads Totais' em 0
olx_pf['Leads Totais'] = olx_pf['Leads Totais'].fillna(0)

# Converte 'Account ID' para tipo inteiro
olx_pf['Account ID'] = olx_pf['Account ID'].astype(int)

# Converte 'Leads Totais' para tipo inteiro
olx_pf['Leads Totais'] = olx_pf['Leads Totais'].astype(int)

# Transformar 'Último Acesso' em datetime
olx_pf['Último Acesso'] = pd.to_datetime(olx_pf['Último Acesso'], format='%d/%m/%Y', errors='coerce')

# Calcular Nova coluna 'Dias sem acesso'
olx_pf['Dias sem acesso'] = (datetime.now() - olx_pf['Último Acesso']).dt.days - 1

# Transforma em formato brasileiro 'Último Acesso'
olx_pf['Último Acesso'] = olx_pf['Último Acesso'].dt.strftime('%d/%m/%Y')


# Ordenar colunas conforme a ordem
olx_ajustado = olx_pf[['Account ID', 'E-mail', 'Nome da Conta', 'Acessos', 'Último Acesso', 'Dias sem acesso', 'Anúncios Ativos', 'Novos Anúncios',
       'Taxa de Uso', 'Média de Imagens','Anúncios sem Fotos','Destaques', 'Chats Respondidos', 'Leads Totais', 'Link do Perfil']]

# Transformar valores NaN de 'Dias sem acesso' em 0
olx_ajustado['Dias sem acesso'] = olx_ajustado['Dias sem acesso'].fillna(0)

# Converte 'Dias sem acesso' para tipo inteiro
olx_ajustado['Dias sem acesso'] = olx_ajustado['Dias sem acesso'].astype(int)

# Remover duplicadas
olx_ajustado = olx_ajustado.drop_duplicates()

# Transformar em arquivo xlsx
olx_ajustado.to_excel('olx_ajustado.xlsx', index=False)
```
