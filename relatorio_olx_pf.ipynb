```python
import pandas as pd
import numpy as np
import re
from datetime import datetime

olx = pd.read_excel('Movida_Vendedores.xlsx')
email_loja = pd.read_excel('email_loja.xlsx')
# propostas = pd.read_csv('propostas.csv', sep=";", low_memory=False)

olx_pf = olx.copy()

# Eliminar colunas desnecessárias
olx_pf = olx_pf.drop(columns=['Grupo','Nome da Conta', 'Validação da Conta',
       'Validação do Telefone', 'Chat ou Form. Email',
       'Login no app',  'Dispositivo Acessado',
       'Início do Plano', 'Plano',
        'Tempo Médio de Resposta', 'Cliques WhatsApp',
       'Cliques Telefone',  'Média de Anúncios',
       'Lead por Anúncio'])

# Junção com planilha email_loja
olx_pf = olx_pf.merge(email_loja, on='E-mail', how='left')

# Transformar valores NaN de 'Account ID' em 0
olx_pf['Account ID'] = olx_pf['Account ID'].fillna(0)

# Transformar valores NaN de 'Leads Totais' em 0
olx_pf['Leads Totais'] = olx_pf['Leads Totais'].fillna(0)

# Converte 'Account ID' para type int
olx_pf['Account ID'] = olx_pf['Account ID'].astype(int)

# Converte 'Leads Totais' para type int
olx_pf['Leads Totais'] = olx_pf['Leads Totais'].astype(int)

# Transformar 'Último Acesso' em datetime
olx_pf['Último Acesso'] = pd.to_datetime(olx_pf['Último Acesso'], format='%d/%m/%Y', errors='coerce')

# Calcular Nova coluna 'Dias sem acesso'
olx_pf['Dias sem acesso'] = (datetime.now() - olx_pf['Último Acesso']).dt.days - 1

# Transforma em formato brasileiro 'Último Acesso'
olx_pf['Último Acesso'] = olx_pf['Último Acesso'].dt.strftime('%d/%m/%Y')


# Ordenar colunas conforme a ordem
olx_ajustado = olx_pf[['Account ID', 'E-mail', 'Nome da Conta', 'Acessos', 'Último Acesso', 'Dias sem acesso', 'Anúncios Ativos', 'Novos Anúncios',
       'Taxa de Uso', 'Média de Imagens','Anúncios sem Fotos','Destaques', 'Chats Respondidos', 'Leads Totais', 'Link do Perfil']].copy()

# Transformar valores NaN de 'Dias sem acesso' em 0 e para type int
olx_ajustado.loc[:, 'Dias sem acesso'] = olx_ajustado['Dias sem acesso'].fillna(0)
olx_ajustado.loc[:, 'Dias sem acesso'] = olx_ajustado['Dias sem acesso'].astype(int)


# Remover duplicadas
olx_ajustado = olx_ajustado.drop_duplicates()

# Transformar em arquivo xlsx
olx_ajustado.to_excel('olx_ajustado.xlsx', index=False)
```

```python

import pandas as pd
from datetime import datetime

# --- Carregando arquivos enviados pela Mobiauto ---
propostas = pd.read_csv("propostas.csv", sep=";", low_memory=False)
email_loja = pd.read_excel("email_loja.xlsx")

# Copiar df para manipulação
mobiauto_olx = propostas.copy()

# Garantir que colunas sejam string
cols_str = ["Origem", "Canal", "Nome PDV"]
mobiauto_olx[cols_str] = mobiauto_olx[cols_str].astype("string")

# --- Filtros iniciais ---
mobiauto_olx = mobiauto_olx[
    (mobiauto_olx["Origem"] == "OLX Pessoa Física")
    & (mobiauto_olx["Canal"] == "Loja")
]

```

```python
# --- Junção com email_loja ---
propostas_ajustadas = propostas.merge(
    email_loja,
    left_on="Nome PDV",
    right_on="Nome da Conta",
    how="left"
)
```

```python


# --- Filtrando apenas as lojas determinadas ---
lojas_permitidas = [
    "Seminovos Movida Niterói",
    "Seminovos Movida Rio de Janeiro Tijuca",
    "Seminovos Movida Juiz de Fora",
    "Seminovos Movida Curitiba Mario Tourinho",
    "Seminovos Movida Curitiba Floriano",
    "Seminovos Movida Itajaí",
    "Seminovos Movida Caxias do Sul",
    "Seminovos Movida Rio de Janeiro Duque de Caxias",
    "Seminovos Movida São José dos Pinhais",
    "Seminovos Movida Maringá",
    "Seminovos Movida Porto Alegre Sarandi",
    "Seminovos Movida Londrina",
    "Seminovos Movida Rio de Janeiro Vila Valqueire",
    "Seminovos Movida Nova Iguaçu",
    "Seminovos Movida Rio de Janeiro Av. das Américas",
    "Seminovos Movida Canoas",
    "Seminovos Movida Rio de Janeiro Campo Grande",
    "Seminovos Movida São José",
    "Seminovos Movida Porto Alegre Ipiranga",
    "Seminovos Movida Rio de Janeiro Barra da Tijuca",
    "Seminovos Movida Joinville"
]

# Filtrar propostas com lojas permitidas
mobiauto_olx = propostas_ajustadas[propostas_ajustadas["Nome da Conta"].isin(lojas_permitidas)]

# --- Transformando e filtrando colunas de datas ---
mobiauto_olx["Data ultima interação"] = pd.to_datetime(
    mobiauto_olx["Data ultima interação"], format="%d/%m/%Y", errors="coerce"
)

# Filtrar apenas dados a partir do início do mês atual
primeiro_dia_mes = datetime.today().replace(day=1)
mobiauto_olx = mobiauto_olx[mobiauto_olx["Data ultima interação"] >= primeiro_dia_mes]
```

```python
# Converter para dd/mm/yyyy
mobiauto_olx["Data ultima interação"] = mobiauto_olx["Data ultima interação"].dt.strftime("%d/%m/%Y")



```
